ฟังก์ชันนี้จะคืนผลรวมของเงินรางวัลใหญ่ที่ได้

และลิสต์ย่อยรางวัลข้างเคียงรางวัลที่ 1

function `first_digits_check` ตรวจสลากกับรางวัลเลขหน้า 3 ตัว และคำนวณเงินรางวัล
ข้อมูลนำเข้า คือ เลขสลากที่ต้องการตรวจรางวัล 1 จำนวน (ในรูปแบบ str)
และลิสต์ย่อยรางวัลเลขหน้า 3 ตัว

function `last_digits_check` ตรวจสลากกับรางวัลเลขท้าย และคำนวณเงินรางวัล
ข้อมูลนำเข้า คือ เลขสลากที่ต้องการตรวจรางวัล 1 จำนวน (ในรูปแบบ `str`)
และลิสต์ย่อยรางวัลเลขท้าย 3 ตัว หรือ 2 ตัว
ฟังก์ชันนี้จะคืนผลรวมของเงินรางวัลเลขท้าย 3 ตัวหรือ 2 ตัว ที่ได้

function `digit_prizes` ตรวจสลากกับรางวัลเลขหน้าและเลขท้ายทั้งหมด และคำนวณเงินรางวัล
ข้อมูลนำเข้า คือ เลขสลากที่ต้องการตรวจรางวัล 1 จำนวน (ในรูปแบบ `str`) และลิสต์ที่ประกอบด้วยลิสต์ย่อยรางวัลเลขหน้า 3 ตัว เลขท้าย 3 ตัว และ เลขท้าย 2 ตัว ตามลำดับ
ฟังก์ชันนี้จะคืนผลรวมของเงินรางวัลเลขหน้าและเลขท้ายทั้งหมดที่ได้


function `total_prize` ตรวจสลากกับทุกรางวัล และ สรุปเงินรางวัลรวม
ข้อมูลนำเข้า คือ เลขสลากที่ต้องการตรวจรางวัล 1 จำนวน (ในรูปแบบ `str`) และลิสต์ `all_prizes`
ฟังก์ชันนี้จะคืนผลรวมของเงินรางวัลที่ได้จากทุกรางวัลที่ออก

````python
# HW03 (ไม่ลบหรือแก้ไขบรรทัดนี้ หรือเพิ่มอะไรก่อนบรรทัดนี้ โดยเด็ดขาด)

# เขียนทั้งโปรแกรมใน cell นี่ cell เดียวเท่านั้น


def prize_check(lot, prize):
    # lot คือหมายเลขสลากที่ต้องการตรวจรางวัล ในรูปแบบ str
    # prize คือ ลิสต์ย่อยรางวัลใหญ่ 1 รางวัล
    # คืนผลรวมของเงินรางวัลที่ได้
    return sum(prize[-1] for i in prize[:-1] if lot == i)


def big_prizes(lot, prizes):
    # lot คือหมายเลขสลากที่ต้องการตรวจรางวัล ในรูปแบบ str
    # prizes คือ ลิสต์ที่ประกอบด้วยลิสต์ย่อยรางวัลที่ 1-5
    # คืนผลรวมของเงินรางวัลที่ได้
    return sum(prize_check(lot, prize) for prize in prizes)


def next_to_first(lot, prize):
    # lot คือหมายเลขสลากที่ต้องการตรวจรางวัล ในรูปแบบ str
    # prize คือลิสต์ย่อยรางวัลข้างเคียงรางวัลที่ 1
    # คืนผลรวมของเงินรางวัลข้างเคียงรางวัลที่1 ที่ได้
    def close(lot, tar):
        tar = int(tar)
        lot = int(lot)
        return (tar+1) % 1000000 == lot or (tar-1) % 1000000 == lot
    return sum(prize[-1] for i in prize[:-1] if close(lot, i))


def first_digits_check(lot, prize):
    # lot คือหมายเลขสลากที่ต้องการตรวจรางวัล ในรูปแบบ str
    # prize คือลิสต์ย่อยรางวัลเลขหน้า 3 ตัว
    # คืนผลรวมของเงินรางวัลที่ได้
    return sum(prize[-1] for i in prize[:-1] if lot[:3] == i)


def last_digits_check(lot, prize):
    # lot คือหมายเลขสลากที่ต้องการตรวจรางวัล ในรูปแบบ str
    # prize คือลิสต์ย่อยรางวัลเลขท้าย 3 ตัว หรือ 2 ตัว
    # คืนผลรวมของเงินรางวัลที่ได้
    # ลิสต์ที่ประกอบด้วยลิสต์ย่อยรางวัลเลขหน้า 3 ตัว เลขท้าย 3 ตัว และ เลขท้าย 2 ตัว ตามลำดับ
    return sum(prize[-1] for i in prize[:-1] if lot[-3:] == i or lot[-2:] == i)


def digit_prizes(lot, prizes):
    # lot คือหมายเลขสลากที่ต้องการตรวจรางวัล ในรูปแบบ str
    # prizes คือลิสต์ที่ประกอบด้วยลิสต์ย่อยรางวัลเลขหน้า 3 ตัว เลขท้าย 3 ตัว และ เลขท้าย 2 ตัว ตามลำดับ
    # คืนผลรวมของเงินรางวัลที่ได้
    return sum([first_digits_check(lot, prizes[0]),
                last_digits_check(lot, prizes[1]),
                last_digits_check(lot, prizes[2])])


def total_prize(lot, all_prizes):
    # lot คือหมายเลขสลากที่ต้องการตรวจรางวัล ในรูปแบบ str
    # all_prizes คือลิสต์ที่ประกอบด้วย "ลิสต์ย่อยรางวัล" ทุกรางวัลตามลำดับในตัวอย่าง
    # คืนผลรวมของเงินรางวัลที่ได้จากทุกรางวัลที่ออก
    return sum([big_prizes(lot, all_prizes[:5]),
               next_to_first(lot, all_prizes[5]),
               digit_prizes(lot, all_prizes[6:])])
````

````python
# ตัวอย่าง

all_prizes = [['555556', 6000000],
              ['111111', '222222', 200000],
              ['333333', '444444', '666666', 80000],
              ['555555', '555555', '777777', '888888', 40000],
              ['000001', '000002', '000003', '000004', '000005', '000006', 20000], 
              ['555556', 100000],
              ['555', '666', 4000],
              ['777', '888', 4000],
              ['55', 2000]
              ]
lot = '555555'
print(prize_check(lot,all_prizes[0]))     # ตรวจรางวัลที่ 1
print(prize_check(lot,all_prizes[3]))     # ตรวจรางวัลที่ 4
print(big_prizes(lot,all_prizes[:5]))     # ตรวจรางวัลใหญ่รวม
print(next_to_first(lot,all_prizes[5]))   # ตรวจรางวัลข้างเคียงรางวัลที่ 1
print(first_digits_check(lot,all_prizes[6]))   # ตรวจรางวัลเลขหน้า 3 ตัว
print(last_digits_check(lot,all_prizes[8]))    # ตรวจรางวัลเลขท้าย 2 ตัว
print(digit_prizes (lot,all_prizes[6:]))  # ตรวจรางวัลเลขหน้าและเลขท้ายรวม
print(total_prize (lot,all_prizes))       # ตรวจรวมทุกรางวัล
````
