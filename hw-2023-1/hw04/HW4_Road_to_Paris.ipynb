````python
# HW04 (ไม่ลบหรือแก้ไขบรรทัดนี้ หรือเพิ่มอะไรก่อนบรรทัดนี้ โดยเด็ดขาด)

# เขียนทั้งโปรแกรมใน cell นี่ cell เดียวเท่านั้น
# จะเขียนฟังก์ชันใด ๆ เพิ่ม ก็เขียนใน cell นี้ตรงบริเวณที่กำหนดไว้ให้

def match_to_dict(match_result_str):
    match_result = {}
    # เติมคำสั่งตรงนี้
    mrs = match_result_str.split()
    t1, t2 = mrs[:2]
    scores = mrs[2:]
    l1, l2 = [t1], [t2]
    for score in scores:
        s1, s2 = score.split('-')
        l1.append(int(s1))
        l2.append(int(s2))
    match_result['team_1'] = l1
    match_result['team_2'] = l2
    return match_result


def calculate_match_stat(match_result):
    match_stat = {}
    # เติมคำสั่งตรงนี้
    t1, t2 = match_result['team_1'][0], match_result['team_2'][0]
    s1, s2 = match_result['team_1'][1:], match_result['team_2'][1:]
    ts1, ts2 = sum(s1), sum(s2)
    w1 = sum(i > j for i, j in zip(s1, s2))
    w2 = sum(i < j for i, j in zip(s1, s2))

    w = w1 > w2

    ms1, ms2 = (3, 0) if min(w1, w2) <= 1 else (2, 1)
    if w1 < w2:
        ms2, ms1 = ms1, ms2

    match_stat[t1] = [int(w), int(not w), ms1, w1, w2, ts1, ts2]
    match_stat[t2] = [int(not w), int(w), ms2, w2, w1, ts2, ts1]
    return match_stat


def update_stat(stat, match_stat):
    # เติมคำสั่งตรงนี้
    for k, v in match_stat.items():
        if (k not in stat):
            stat[k] = v
        else:
            stat[k] = [v[i] + stat[k][i] for i in range(len(v))]
    return stat


def load_matches(filename):
    matches = []
    # เติมคำสั่งตรงนี้
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()
        for line in lines:
            matches.append(match_to_dict(line))

    return matches
# ----------------------------------------------
# เพิ่มฟังก์ชันที่เขียนเพิ่มเองบริเวณนนี้
# ----------------------------------------------


# ----------------------------------------------
# สิ้นสุดบริเวณที่เขียนฟังก์ชันเพิ่มเติมเอง
# ----------------------------------------------

# ----------------------------------------------
# ห้ามแก้ไขโปรแกรมถัดจากบรรทัดนี้
# ----------------------------------------------
def stat_to_rank(stat):
    '''
    stat: {teamname: [won, loss, point, set_won, set_loss, point_won, point_loss]}
    '''
    temp = [[-wn, -pnt, -set_ratio(stwn, stlss), -ratio((pntwn/pntlss)), team, lss, stwn, stlss, pntwn, pntlss]
            for team, [wn, lss, pnt, stwn, stlss, pntwn, pntlss] in stat.items()]
    temp.sort()
    return [[team, -wn, lss, -pnt, -strt, -pntrt, stwn, stlss, pntwn, pntlss] for [wn, pnt, strt, pntrt, team, lss, stwn, stlss, pntwn, pntlss] in temp]


def set_ratio(won, loss):
    if loss == 0:
        return float('inf')
    else:
        return ratio(won/loss)


def ratio(value):
    return round(int(value*1000)/1000, 3)


def show_result(pool, table):
    pool = ' '.join(pool.split('.')[0].split('_'))
    k = (21-len(pool))//2 - 1
    print(f'{"#"*k} {pool} {"#"*k} | Won- | Point |  Set  | Point |   Set-   |  Point-  |')
    print('   Team              | Loss |       | Ratio | Ratio | Won-Loss | Won-Loss |')
    for i, t in enumerate(table):
        set_ratio = "MAX" if t[4] == float("inf") else t[4]
        format = '^5.3f' if set_ratio != 'MAX' else '^5'
        set_ratio_out = f'{set_ratio:{format}}'
        print(
            f'{i+1:2} {t[0]:20}{t[1]}-{t[2]}{t[3]:7}     {set_ratio_out}   {t[5]:^5.3f}     {t[6]:>2}-{t[7]:<2}   {t[8]:4}-{t[9]:3}')
    print('-'*75)


pools = ['Pool_A.txt', 'Pool_B.txt', 'Pool_C.txt']
for pool in pools:
    matches = load_matches(pool)
    stat = {}
    for match_result in matches:
        update_stat(stat, calculate_match_stat(match_result))
    teams = stat_to_rank(stat)
    show_result(pool, teams)
````

Run code cell นี้ทุกครั้งก่อน run กรณีทดสอบ


````python
def test(expected, result):
    bold_red = '\033[1;91m'
    bold_green = '\033[1;92m'
    normal = '\033[0m'
    print('Result:', bold_green if expected ==
          result else bold_red, expected == result, normal)
    print('Expected:')
    print(expected)
    print('Yours:')
    print(result)
    print('-'*20)
````
