- นิสิตสามารถลบ doc_str (คอมเม้นที่ตามหลังฟังก์ชันที่ปิดหัวท้ายด้วย """) ที่โจทย์ระบุไว้ในแต่ละฟังก์ขันได้ แต่ถ้าไม่อยากลบ ก็ให้เก็บ doc_str ไว้ที่เดิมหลังบรรทัด def และถ้าจะเพิ่ม code ใหม่ในฟังก์ชัน ให้เยื้องให้อยู่ในแนวเดียวกับ doc_str


```python
# HW5_Quiz_Analysis
def get_sid_point(stu_rec):
    """
    รับ stu_rec ซึ่งเป็นข้อมูลประเภทสตริง แทนข้อมูลคะแนนแต่ละครั้งในเกรดเดอร์
    ซึ่งจะประกอบด้วย
    - เลขประจำตัวนิสิต (แต่ในการบ้านนี้จะใช้นามแฝงแทนรหัสนิสิตจริงๆ)
    - คะแนน (คะแนนจะให้ค่ามาได้ตั้งแต่ 0-100)
    คั่นด้วยเครื่องหมาย ","

    ให้คืนค่าเป็น tuple ที่ประกอบด้วยสมาชิกสองตัว
    - ตัวแรก เลขประจำตัวนิสิต เป็นข้อมูลประเภทสตริง
    - ตัวที่สอง จะเป็นคะแนน เป็นข้อมูลประเภทจำนวนเต็ม

    ตัวอย่าง
    ให้ stu_rec เป็น "MrBond,20"
    ถ้าเรียก get_sid_section(stu_rec) ก็จะคืนค่าเป็น ("MrBond", 20) <-- โปรดสังเกตว่าเป็น tuple
    ให้ stu_rec เป็น "stu_CompProg,0"
    ถ้าเรียก get_sid_section(stu_rec) ก็จะคืนค่าเป็น ("stu_CompProg", 0)
    """
    return stu_rec.split(',')[0], int(stu_rec.split(',')[1])


def get_stu_max_point(stu_recs):
    """
    รับ stu_recs ซึ่งเป็นข้อมูลประเภทลิสต์ แทนข้อมูลคะแนนทั้งหมดของโจทย์เกรดเดอร์หนึ่งข้อ
    โดยสมาชิกแต่ละตัวจะเป็นข้อมูลประเภทสตริงที่อยู่ในแบบเดียวกับ argument stu_rec ของฟังก์ชัน
    get_sid_point
    โดยคะแนนของนิสิตเลขประจำตัวเดียวกันอาจจะมีได้หลายครั้งใน stu_recs

    ให้คืนค่าเป็น dict ที่มีคุณสมบัติดังนี้
    - มี key เป็นเลขประจำตัวนิสิต เป็นข้อมูลประเภทสตริง
    - มี value เป็นคะแนนที่มากที่สุดที่ได้ในโจทย์ข้อนี้

    ตัวอย่าง
    ให้ stu_recs เป็น ["MrBond,20", "stu_CompProg,0", "MrBond,50"]
    ถ้าเรียก get_stu_max_point(stu_recs) ก็จะคืนค่าเป็น
    {'MrBond': 50, 'stu_CompProg': 0}

    คำแนะนำ
    ให้ใช้ประโยชน์จากฟังก์ชัน get_sid_point
    """
    r = dict()
    for stu_rec in stu_recs:
        sid, point = get_sid_point(stu_rec)
        if sid not in r:
            r[sid] = point
        else:
            r[sid] = max(r[sid], point)
    return r


def get_stu_max_points(stu_recss):
    """
    รับ stu_recss ซึ่งเป็นข้อมูลประเภทลิสต์ แทนข้อมูลคะแนนทั้งหมดของโจทย์
    เกรดเดอร์หลายข้อ โดยสมาชิกแต่ละตัวจะเป็นข้อมูลประเภท ลิสต์ที่อยู่ใน
    แบบเดียวกับ argument stu_recs ของฟังก์ชัน get_stu_max_point
    โดยจำนวนสมาชิกของลิสต์ จะเท่ากับจำนวนโจทย์ (หนึ่งอันแทนโจทย์หนึ่งข้อ)
    จำนวนสมาชิกจะมีมากกว่า 1 เสมอ และทุกโจทย์จะมีนิสิตส่งอย่างน้อย 1 คนเสมอ
    (อาจจะได้ 0 หรือได้คะแนนเท่าไหร่ก็ได้)

    ให้คืนค่าเป็น dict ที่มีคุณสมบัติดังนี้
    - มี key เป็นเลขประจำตัวนิสิต เป็นข้อมูลประเภทสตริง
    - มี value เป็น ลิสต์ แทนคะแนนที่มากที่สุดที่นิสิตคนดังกล่าวได้ในแต่ละข้อ
      - จำนวนสมาชิกในลิสต์จะเท่ากับจำนวนโจทย์
        - ลำดับในลิสต์จะเรียงตามลำดับโจทย์ที่รับเข้ามา
      - หากมีการส่งในบางโจทย์ บางโจทย์ไม่ได้ส่ง โจทย์ที่ไม่ได้ส่งจะได้ 0

    ตัวอย่าง
    ให้ stu_recss เป็น [["MrBond,20", "stu_CompProg,0", "MrBond,50"],
                      ["stu_CompProg,100", "MrBond,100"]]
    ถ้าเรียก get_stu_max_points(stu_recss) ก็จะคืนค่าเป็น
    {'MrBond': [50, 100], 'stu_CompProg': [0, 100]}
    ให้ stu_recss เป็น [["MrBond,20", "stu_CompProg,0", "MrBond,50"],
                      ["stu_CompProg,100", "Luffy,100", "MrBond,100"],
                      ["MrBond,50", "MrBond,0"]]
    ถ้าเรียก get_stu_max_points(stu_recss) ก็จะคืนค่าเป็น
    {'MrBond': [50, 100, 50], 'stu_CompProg': [0, 100, 0], 'Luffy': [0, 100, 0]}

    คำแนะนำ
    ให้ใช้ประโยชน์จากฟังก์ชัน get_stu_max_point
    """
    r = dict()
    for i, stu_recs in enumerate(stu_recss):
        mp = get_stu_max_point(stu_recs)
        for k, v in mp.items():
            if k not in r:
                r[k] = [0] * i
            r[k].append(v)
        for k, v in r.items():
            if len(r[k]) != i+1:
                r[k].append(0)
    return r


def get_stu_quiz_point(stu_recss):
    """
    รับ stu_recss ซึ่งเป็นข้อมูลประเภทลิสต์ที่อยู่ในแบบเดียวกับ argument stu_recss ของฟังก์ชัน
    get_stu_max_points

    สำหรับฟังก์ชันนี้ stu_recss จะแทนคะแนนของนิสิตในการทำโจทย์ควิซ
    ซึ่งจะมีมากกว่า 1 ข้อเสมอ

    ให้คืนค่าเป็น dict ที่มีคุณสมบัติดังนี้
    - มี key เป็นเลขประจำตัวนิสิต เป็นข้อมูลประเภทสตริง
    - มี value เป็นคะแนนที่ได้จากการทำควิซ เป็นข้อมูลประเภทจำนวนเต็ม
      - คะแนนที่ได้ของนิสิตแต่ละคนจะเป็นคะแนนรวม + คะแนนข้อที่มากที่สุด
      - สมมุติว่าการสอบครั้งนั้นมีควิซ 4 ข้อและนิสิตได้คะแนน 50,0,90,80
        คะแนนที่นิสิตได้จะเป็น 50+0+90+80+90 -> 310
      - หากมีการส่งในบางโจทย์ บางโจทย์ไม่ได้ส่ง โจทย์ที่ไม่ได้ส่งให้ถือว่าได้ 0

    ตัวอย่าง
    ให้ stu_recss เป็น [["MrBond,20", "stu_CompProg,0", "MrBond,50"],
                      ["stu_CompProg,100", "MrBond,100"]]
    ถ้าเรียก get_stu_quiz_point(stu_recss) ก็จะคืนค่าเป็น
    {'MrBond': 250, 'stu_CompProg': 200}
    ให้ stu_recss เป็น [["MrBond,20", "stu_CompProg,0", "MrBond,50"],
                      ["stu_CompProg,100", "Luffy,100", "MrBond,100"],
                      ["MrBond,50", "MrBond,0"]]
    ถ้าเรียก get_stu_quiz_point(stu_recss) ก็จะคืนค่าเป็น
    {'MrBond': 300, 'stu_CompProg': 200, 'Luffy': 200}
    ให้ stu_recss เป็น [["MrBond,20", "stu_CompProg,0", "MrBond,50"],
                      ["stu_CompProg,100", "Luffy,100", "MrBond,100"],
                      ["MrBond,0", "Luffy,100", "Luffy,0", "MrBond,50"],
                      ["Luffy,50"]]
    ถ้าเรียก get_stu_quiz_point(stu_recss) ก็จะคืนค่าเป็น
    {'MrBond': 300, 'stu_CompProg': 200, 'Luffy': 350}

    คำแนะนำ
    ให้ใช้ประโยชน์จากฟังก์ชัน get_stu_max_points
    """
    r, dm = dict(), dict()
    for stu_recs in stu_recss:
        for k, v in get_stu_max_point(stu_recs).items():
            if k in dm:
                dm[k] = max(dm[k], v)
            else:
                dm[k] = v

            r[k] = r.get(k, 0) + v
    for k, v in dm.items():
        r[k] += v
    return r


def get_stu_completed_problem(stu_recss):
    """
    รับ stu_recss ซึ่งเป็นข้อมูลประเภทลิสต์ที่อยู่ในแบบเดียวกับ argument stu_recss ของฟังก์ชัน
    get_stu_max_points

    สำหรับฟังก์ชันนี้ stu_recss จะแทนคะแนนของนิสิตในการทำโจทย์เกรดเดอร์ใดๆ
    ซึ่งจะมีอย่างน้อย 1 ข้อเสมอ

    ให้คืนค่าเป็น dict ที่มีคุณสมบัติดังนี้
    - มี key เป็นเลขประจำตัวนิสิต เป็นข้อมูลประเภทสตริง
    - มี value เป็นจำนวนข้อที่นิสิตได้เต็ม (100 คะแนน) เป็นข้อมูลประเภทจำนวนเต็ม

    ตัวอย่าง
    ให้ stu_recss เป็น [["MrBond,80", "stu_CompProg,100", "MrBond,50", "Lalisa,100"],
                      ["stu_CompProg,100", "MrBond,100", "Lalisa,100"]]
    ถ้าเรียก get_stu_completed_problem(stu_recss) ก็จะคืนค่าเป็น
    {'MrBond': 1, 'stu_CompProg': 2, 'Lalisa': 2}
    ให้ stu_recss เป็น [["101,20", "Lalisa,100"],
                      ["stu_CompProg,100", "Lalisa,100", "MrBond,100"],
                      ["MrBond,50", "Lalisa,100", "MrBond,0"],
                      ["Lalisa,100"],
                      ["Lalisa,100"]]
    ถ้าเรียก get_stu_completed_problem(stu_recss) ก็จะคืนค่าเป็น
    {'101': 0, 'Lalisa': 5, 'stu_CompProg': 1, 'MrBond': 1}

    คำแนะนำ
    ให้ใช้ประโยชน์จากฟังก์ชัน get_stu_max_points
    """
    r = dict()
    for stu_recs in stu_recss:
        for k, v in get_stu_max_point(stu_recs).items():
            if k not in r:
                r[k] = 0
            if v == 100:
                r[k] = r.get(k, 0) + 1
    return r


def get_stu_stat(stu_recss1, stu_recss2):
    """
    รับ stu_recss1 และ stu_recss2 โดย
    - stu_recss1 ซึ่งเป็นข้อมูลประเภทลิสต์ที่อยู่ในแบบเดียวกับ argument stu_recss ของฟังก์ชัน
    get_stu_quiz_point ซึ่งเก็บข้อมูลของโจทย์ที่ใช้ทำตอนสอบ
    - stu_recss2 ซึ่งเป็นข้อมูลประเภทลิสต์ที่อยู่ในแบบเดียวกับ argument stu_recss ของฟังก์ชัน
    get_stu_completed_problem ซึ่งเก็บข้อมูลของโจทย์ที่ทำเป็นแบบฝึกหัด

    ให้คืนค่าเป็น dict ที่มีคุณสมบัติดังนี้
    - มี key เป็นเลขประจำตัวนิสิต เป็นข้อมูลประเภทสตริง
    - มี value เป็น tuple ที่มีสมาชิก 2 ตัว
      - ตัวแรกเป็นคะแนนควิซที่นิสิตทำได้ที่เกิดจากการคำนวณโดยใช้ stu_recss1
        เป็นข้อมูลประเภทจำนวนเต็ม
      - ตัวที่สองเป็นจำนวนโจทย์ที่นิสิตทำได้เต็มที่เกิดจากการคำนวณโดยใช้ stu_recss2
        เป็นข้อมูลประเภทจำนวนเต็ม
      - หากนิสิตมีข้อมูลแค่อย่างใดอย่างหนึ่ง ให้อีกอย่างเป็น 0
        (สามารถมีกรณีที่นิสิตไม่เคยทำโจทย์ได้คะแนนเต็มเลยหรือไม่เคยทำโจทย์เลย แต่มีคะแนนควิซ
        และในทางตรงกันข้าม มีนิสิตที่เคยทำโจทย์ แต่อาจจะไม่เคยส่งคะแนนควิซเลย)

    ตัวอย่าง
    ให้ stu_recss1 เป็น [["MrBond,20", "stu_CompProg,0", "MrBond,50"],
                      ["stu_CompProg,100", "MrBond,100"]]
    ให้ stu_recss2 เป็น [["MrBond,80", "stu_CompProg,100", "MrBond,50"],
                      ["stu_CompProg,100", "MrBond,100"]]
    ถ้าเรียก get_stu_stat(stu_recss1, stu_recss2) ก็จะคืนค่าเป็น
    {'stu_CompProg': (200, 2), 'MrBond': (250, 1)}
    ให้ stu_recss1 เป็น [["MrBond,20", "stu_CompProg,0", "MrBond,50"],
                      ["stu_CompProg,100", "Luffy,100", "MrBond,100"],
                      ["MrBond,50", "MrBond,0"]]
    ให้ stu_recss2 เป็น [["101,20", "Lalisa,100"],
                      ["stu_CompProg,100", "Lalisa,100", "MrBond,100"],
                      ["MrBond,50", "Lalisa,100", "MrBond,0"],
                      ["Lalisa,100"],
                      ["Lalisa,100"]]
    ถ้าเรียก get_stu_stat(stu_recss1, stu_recss2) ก็จะคืนค่าเป็น
    {'MrBond': (300, 1), '101': (0, 0), 'Lalisa': (0, 5), 'stu_CompProg': (200, 1), 'Luffy': (200, 0)}

    คำแนะนำ
    ให้ใช้ประโยชน์จากฟังก์ชัน get_stu_quiz_point และ get_stu_completed_problem
    """
    r = dict()
    qp = get_stu_quiz_point(stu_recss1)
    cp = get_stu_completed_problem(stu_recss2)
    for k, v in qp.items():
        r[k] = (v, cp.get(k, 0))
    for k, v in cp.items():
        r[k] = (qp.get(k, 0), v)

    return r
```

```python
import copy


def test_f(case_name, f_name, expected, f_kwargs, allowed_unsorted_result=False):
    def show_warning(case_name, str1, str2, expected, actual):
        print("WARNING!! Something went wrong at", case_name)
        warning_txt = " ".join(
            [str(e) for e in [str1, str2, expected, "instead of", actual]])
        if len(warning_txt) < 100:
            print(warning_txt)
        else:
            print(str1)
            print(str2)
            print(expected)
            print("instead of")
            print(actual)

    f = eval(f_name)
    mut_val_b4_test = {}
    for key, val in f_kwargs.items():
        if type(val) is list:
            mut_val_b4_test[key] = copy.deepcopy(val)
    call_txt = f_name + "(" + str(list(f_kwargs.values()))[1:-1] + ")"
    result = f(**f_kwargs)
    if allowed_unsorted_result:
        if type(result) is list:
            result.sort()
            expected.sort()
    for key in mut_val_b4_test:
        if mut_val_b4_test[key] != f_kwargs[key]:
            show_warning(case_name,
                         "After calling " + call_txt + ",",
                         "argument " + key + " must have its value as before:",
                         mut_val_b4_test[key],
                         f_kwargs[key])
            return 1
    if type(expected) is not type(result):
        show_warning(case_name, call_txt, "should return variable type", type(
            expected), type(result))
        return 1
    if result != expected:
        show_warning(case_name, call_txt, "should return", expected, result)
        return 1
    return 0


def run_test(f_name, test_cases):
    print("------------------------------------------------------------")
    err_code = 0
    for idx, test_kwargs in enumerate(test_cases):
        case_name = "test " + f_name + " #" + str(idx+1)
        err_code |= test_f(case_name, f_name, **test_kwargs)
    if not err_code:
        print("Your function " + f_name + " is working well so far")


def test_get_sid_point():
    test_cases = []
    test_cases.append({"f_kwargs": {"stu_rec": "MrBond,20"},
                       "expected": ("MrBond", 20)})
    test_cases.append({"f_kwargs": {"stu_rec": "stu_CompProg,0"},
                       "expected": ("stu_CompProg", 0)})
    run_test("get_sid_point", test_cases)


def test_get_stu_max_point():
    test_cases = []
    test_cases.append({"f_kwargs": {"stu_recs": ["MrBond,20", "stu_CompProg,0", "MrBond,50"]},
                       "expected": {'stu_CompProg': 0, 'MrBond': 50}})
    run_test("get_stu_max_point", test_cases)


def test_get_stu_max_points():
    test_cases = []
    test_cases.append({"f_kwargs": {"stu_recss": [["MrBond,20", "stu_CompProg,0", "MrBond,50"],
                                                  ["stu_CompProg,100", "MrBond,100"]]},
                       "expected": {'MrBond': [50, 100], 'stu_CompProg': [0, 100]}})
    test_cases.append({"f_kwargs": {"stu_recss": [["MrBond,20", "stu_CompProg,0", "MrBond,50"],
                                                  ["stu_CompProg,100", "Luffy,100",
                                                      "Luffy,0", "MrBond,100"],
                                                  ["MrBond,0", "MrBond,50"]]},
                       "expected": {'MrBond': [50, 100, 50], 'stu_CompProg': [0, 100, 0], 'Luffy': [0, 100, 0]}})
    run_test("get_stu_max_points", test_cases)


def test_get_stu_quiz_point():
    test_cases = []
    test_cases.append({"f_kwargs": {"stu_recss": [["MrBond,20", "stu_CompProg,0", "MrBond,50"],
                                                  ["stu_CompProg,100", "MrBond,100"]]},
                       "expected": {'MrBond': 250, 'stu_CompProg': 200}})
    test_cases.append({"f_kwargs": {"stu_recss": [["MrBond,20", "stu_CompProg,0", "MrBond,50"],
                                                  ["stu_CompProg,100", "Luffy,100",
                                                      "Luffy,0", "MrBond,100"],
                                                  ["MrBond,0", "MrBond,50"]]},
                       "expected": {'MrBond': 300, 'stu_CompProg': 200, 'Luffy': 200}})
    test_cases.append({"f_kwargs": {"stu_recss": [["MrBond,20", "stu_CompProg,0", "MrBond,50"],
                                                  ["stu_CompProg,100",
                                                      "Luffy,100", "MrBond,100"],
                                                  ["MrBond,0", "Luffy,100",
                                                      "Luffy,0", "MrBond,50"],
                                                  ["Luffy,50"]]},
                       "expected": {'MrBond': 300, 'stu_CompProg': 200, 'Luffy': 350}})
    run_test("get_stu_quiz_point", test_cases)


def test_get_stu_completed_problem():
    test_cases = []
    test_cases.append({"f_kwargs": {"stu_recss": [["MrBond,80", "stu_CompProg,100", "MrBond,50", "Lalisa,100"],
                                                  ["stu_CompProg,100", "MrBond,100", "Lalisa,100"]]},
                       "expected": {'MrBond': 1, 'stu_CompProg': 2, 'Lalisa': 2}})
    test_cases.append({"f_kwargs": {"stu_recss": [["101,20", "Lalisa,100"],
                                                  ["stu_CompProg,100",
                                                      "Lalisa,100", "MrBond,100"],
                                                  ["MrBond,50", "Lalisa,100",
                                                      "MrBond,0"],
                                                  ["Lalisa,100"],
                                                  ["Lalisa,100"]]},
                       "expected": {'101': 0, 'Lalisa': 5, 'stu_CompProg': 1, 'MrBond': 1}})
    test_cases.append({"f_kwargs": {"stu_recss": [["101,20", "Lalisa,90"],
                                                  ["Lalisa,100", "Lalisa,20", "101,90", "101,10"]]},
                       "expected": {'101': 0, 'Lalisa': 1}})
    run_test("get_stu_completed_problem", test_cases)


def test_get_stu_stat():
    test_cases = []
    test_cases.append({"f_kwargs": {"stu_recss1": [["MrBond,20", "stu_CompProg,0", "MrBond,50"],
                                                   ["stu_CompProg,100", "MrBond,100"]],
                                    "stu_recss2": [["MrBond,80", "stu_CompProg,100", "MrBond,50"],
                                                   ["stu_CompProg,100", "MrBond,100"]]},
                       "expected": {'stu_CompProg': (200, 2), 'MrBond': (250, 1)}})
    test_cases.append({"f_kwargs": {"stu_recss1": [["MrBond,20", "stu_CompProg,0", "MrBond,50"],
                                                   ["stu_CompProg,100",
                                                       "Luffy,100", "MrBond,100"],
                                                   ["MrBond,50", "MrBond,0"]],
                                    "stu_recss2": [["101,20", "Lalisa,100"],
                                                   ["stu_CompProg,100",
                                                       "Lalisa,100", "MrBond,100"],
                                                   ["MrBond,50", "Lalisa,100",
                                                       "MrBond,0"],
                                                   ["Lalisa,100"],
                                                   ["Lalisa,100"]]},
                       "expected": {'MrBond': (300, 1),
                                    '101': (0, 0),
                                    'Lalisa': (0, 5),
                                    'stu_CompProg': (200, 1),
                                    'Luffy': (200, 0)}})
    test_cases.append({"f_kwargs": {"stu_recss1": [["MrBond,20", "stu_CompProg,0", "MrBond,50"],
                                                   ["stu_CompProg,100",
                                                       "Luffy,100", "MrBond,100"],
                                                   ["MrBond,50", "MrBond,0"]],
                                    "stu_recss2": [["101,20", "Lalisa,90"],
                                                   ["Lalisa,100", "Lalisa,20", "101,90", "101,10"]]},
                       "expected": {'Lalisa': (0, 1),
                                    'stu_CompProg': (200, 0),
                                    'MrBond': (300, 0),
                                    'Luffy': (200, 0),
                                    '101': (0, 0)}})
    run_test("get_stu_stat", test_cases)


test_get_sid_point()
test_get_stu_max_point()
test_get_stu_max_points()
test_get_stu_quiz_point()
test_get_stu_completed_problem()
test_get_stu_stat()
```
